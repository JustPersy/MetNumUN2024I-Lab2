# Análisis de la Base de Datos 'Red Unidos'
Este análisis explora datos relacionados con la Red Unidos, como género, edad, nivel educativo y discapacidad.

### Instrucciones:
1. Asegúrese de que el archivo `BD Mpios busqueda activa_Red Unidos.xlsx` esté cargado en el entorno de Google Colab.
2. Si el archivo está en la carpeta predeterminada de Colab, actualice la ruta según corresponda.

```python
# Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt

# Especificar la ruta del archivo
# Si el archivo está en Colab, utilice: '/content/sample_data/BD Mpios busqueda activa_Red Unidos.xlsx'
file_path = '/content/sample_data/BD Mpios busqueda activa_Red Unidos.xlsx'

# Leer los datos del archivo Excel
data = pd.read_excel(file_path, sheet_name='RED UNIDOS')
data.head()
```

## Selección y Limpieza de Datos
Se seleccionan las columnas más relevantes para este análisis:

```python
# Selección de columnas relevantes
selected_data = data[['MUNICIPIO', 'SEXO', 'EDAD', 'NIVELEDUCATIVOMASALTOALCANZADO', 'DISCAPACIDADEDUCACIONESPECIAL']]

# Limpiar datos eliminando filas con valores faltantes en las columnas clave
cleaned_data = selected_data.dropna(subset=['SEXO', 'EDAD', 'NIVELEDUCATIVOMASALTOALCANZADO'])
cleaned_data.head()
```

## Visualización de Datos

```python
# Distribución por Género
gender_counts = cleaned_data['SEXO'].value_counts()
gender_counts.plot(kind='bar', title='Distribución por Género', ylabel='Cantidad', xlabel='Género')
plt.show()
```

```python
# Distribución de Edades
cleaned_data['EDAD'].plot(kind='hist', bins=20, title='Distribución de Edades', ylabel='Cantidad', xlabel='Edad')
plt.show()
```

```python
# Nivel Educativo Más Alto Alcanzado
education_counts = cleaned_data['NIVELEDUCATIVOMASALTOALCANZADO'].value_counts()
education_counts.plot(kind='bar', title='Nivel Educativo Más Alto Alcanzado', ylabel='Cantidad', xlabel='Nivel Educativo')
plt.show()
```

```python
# Discapacidad o Educación Especial
disability_counts = cleaned_data['DISCAPACIDADEDUCACIONESPECIAL'].value_counts()
disability_counts.plot(kind='pie', autopct='%1.1f%%', title='Discapacidad o Educación Especial')
plt.ylabel('')
plt.show()
```

## Análisis e Interpretación
1. **Distribución por género**: El género masculino tiene una leve predominancia.
2. **Distribución de edades**: La mayoría se concentra en rangos jóvenes-adultos.
3. **Nivel educativo**: Predominan los niveles básicos (primaria y secundaria).
4. **Discapacidad o educación especial**: La mayoría no reporta condiciones especiales.
